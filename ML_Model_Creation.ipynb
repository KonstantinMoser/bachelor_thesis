```python
from sklearn.preprocessing import StandardScaler
#Feature scaling
scaler= StandardScaler().fit(normal_db_train)
# Transform training set
normal_db_train_scaled = scaler.transform(normal_db_train)
# Transform test set
normal_db_test_scaled = scaler.transform(normal_db_test)
# Transform diffrent phases
phase1_scaled = scaler.transform(phase1)
phase2_scaled = scaler.transform(phase2)
phase3_scaled = scaler.transform(phase3)
phase4_scaled = scaler.transform(phase4)
phase5_scaled = scaler.transform(phase5)


def accuracy_normal(unique, counts):
  print(unique," ",counts)
  if len(unique) == 1:
    if unique[0] == 1:
      return 0
    if unique[1] == 0:
      return 1
  else:
    per = counts[0]/(counts[0]+counts[1])*100
    return round(per, dec)

def accuracy_attack(unique, counts):
  print(unique," ",counts)
  if len(unique) == 1:
    if unique[0] == 1:
      return 1
    if unique[1] == 0:
      return 0 
  else:
    per = counts[1]/(counts[0]+counts[1])*100
    return round(per, dec)


#def precision(tp, fp):
#  return (tp / (tp+fp))
#def recall(tp, fn):
#  return (tp / (tp+fn))
#def f1score(pr, r):
#  return 2*(pr*r / (pr+r))
#def accuracy(tp, fp, fn, tn):
#  return (tp + tn)/(tp+fp+fn+tn)
```

```python
#Supervised Machine Learning

#Define the Labels
normal_db['label'] = 0
phase1['label'] = 1
phase2['label'] = 2
phase3['label'] = 3
phase4['label'] = 4
phase5['label'] = 5

#Define Feature Columns
feature_cols = ['block:block_bio_backmerge', 'block:block_bio_remap',
       'block:block_dirty_buffer', 'block:block_getrq',
       'block:block_touch_buffer', 'block:block_unplug', 'clk:clk_set_rate',
       'cpu-migrations', 'cs', 'fib:fib_table_lookup',
       'filemap:mm_filemap_add_to_page_cache', 'gpio:gpio_value',
       'ipi:ipi_raise', 'irq:irq_handler_entry', 'irq:softirq_entry',
       'jbd2:jbd2_handle_start', 'jbd2:jbd2_start_commit', 'kmem:kfree',
       'kmem:kmalloc', 'kmem:kmem_cache_alloc', 'kmem:kmem_cache_free',
       'kmem:mm_page_alloc', 'kmem:mm_page_alloc_zone_locked',
       'kmem:mm_page_free', 'kmem:mm_page_pcpu_drain', 'mmc:mmc_request_start',
       'net:net_dev_queue', 'net:net_dev_xmit', 'net:netif_rx', 'page-faults',
       'pagemap:mm_lru_insertion', 'preemptirq:irq_enable',
       'qdisc:qdisc_dequeue', 'random:get_random_bytes',
       'random:mix_pool_bytes_nolock', 'random:urandom_read',
       'raw_syscalls:sys_enter', 'raw_syscalls:sys_exit', 'rpm:rpm_resume',
       'rpm:rpm_suspend', 'sched:sched_process_exec',
       'sched:sched_process_free', 'sched:sched_process_wait',
       'sched:sched_switch', 'sched:sched_wakeup', 'signal:signal_deliver',
       'signal:signal_generate', 'skb:consume_skb', 'skb:kfree_skb',
       'skb:skb_copy_datagram_iovec', 'sock:inet_sock_set_state',
       'task:task_newtask', 'tcp:tcp_destroy_sock', 'tcp:tcp_probe',
       'timer:hrtimer_start', 'timer:timer_start',
       'workqueue:workqueue_activate_work', 'writeback:global_dirty_state',
       'writeback:sb_clear_inode_writeback', 'writeback:wbc_writepage',
       'writeback:writeback_dirty_inode',
       'writeback:writeback_dirty_inode_enqueue',
       'writeback:writeback_dirty_page',
       'writeback:writeback_mark_inode_dirty',
       'writeback:writeback_pages_written', 'writeback:writeback_single_inode',
       'writeback:writeback_write_inode', 'writeback:writeback_written']

#Split all Datasets
shuffle_bool = False
normal_db_train, normal_db_test = train_test_split(normal_db, test_size=0.20, random_state=state, shuffle=shuffle_bool)
phase1_train, phase1_test = train_test_split(phase1, test_size=0.20, random_state=state, shuffle=shuffle_bool)
phase2_train, phase2_test = train_test_split(phase2, test_size=0.20, random_state=state, shuffle=shuffle_bool)
phase3_train, phase3_test = train_test_split(phase3, test_size=0.20, random_state=state, shuffle=shuffle_bool)
phase4_train, phase4_test = train_test_split(phase4, test_size=0.20, random_state=state, shuffle=shuffle_bool)
phase5_train, phase5_test = train_test_split(phase5, test_size=0.20, random_state=state, shuffle=shuffle_bool)

#All Training Datasets together
db_train = pd.concat([normal_db_train, phase1_train, phase2_train, phase3_train, phase4_train, phase5_train])
X_train = db_train[feature_cols] 
y_train = db_train.label 

#All Testing Datasets together
all_test = pd.concat([normal_db_test, phase1_test, phase2_test, phase3_test, phase4_test, phase5_test])
X_test = all_test[feature_cols] # Features
y_test = all_test.label # Target variable

#Scaling
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler().fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Decision Tree
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

y_pred_all=clf.predict(X_test)
print("Accuracy All:", round(metrics.accuracy_score(y_test, y_pred_all)*100, dec), "%")
from sklearn.metrics import classification_report, confusion_matrix
cm = confusion_matrix(y_test, y_pred_all)
print(cm)
print(cm.diagonal()/cm.sum(axis=1))
print(classification_report(y_test, y_pred_all))
```
